```python
sequence = "ATGCCGTAGCTAAGGCTT"
pos = 4   # delete base at index 4
mutated = sequence[:pos] + sequence[pos+1:]
print("Original:", sequence)
print("deletion Mutation :", mutated)
```

```python
sequence='ATGCCGTAGCTAAGGCTT'
substitution_Mutation='ATGCAGTAGCTAAGGCTT'
insertion_Mutation='ATGCACGTAGCTAAGGCTT'
deletion_Mutation ='ATGCGTAGCTAAGGCTT'
print('original length:',len(sequence))
print('substitution_Mutation:' ,len(substitution_Mutation))
print('insertion_Mutation:' ,len(insertion_Mutation))
print('deletion Mutation :',len(deletion_Mutation ))
```

```python
#substitution mutation
length=min(len(sequence),len(substitution_Mutation))
matches=0

for i in range(length):
    if sequence[i]==substitution_Mutation[i]:
        matches+=1
identity_percentage=(matches/length)*100
print(f"identity percentage:{identity_percentage:.2f}%")
```
