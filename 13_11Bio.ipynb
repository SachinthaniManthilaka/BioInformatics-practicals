```python
sequence1 = 'ATGCATAG'
sequence2 = 'ATGAATGG'


total_length=len(sequence1)
exact_matches = 0

for i in range(total_length):
    if sequence1[i] == sequence2[i]:
        exact_matches += 1

identity_percentage = (exact_matches / total_length) * 100
print(f"Identity Percentage: ",identity_percentage,"%")
```

```python
sequence1='ATGGCTGAACTG'
sequence2='ATGGCCGAATTG'

total_length=len(sequence1)
exact_matches = 0

complement = {"A":"T", "T":"A", "C":"G", "G":"C"}


complements_seq1= ""
for base in sequence1:
    complements_seq1 += complement[base]
    
complements_seq2= ""
for base in sequence2:
    complements_seq2 += complement[base]



mRNA = {"A":"A", "T":"U", "C":"C", "G":"G"}
mRNA_change1=""
for base in complements_seq1:
    mRNA_change1 += mRNA[base]
    
mRNA_change2=""
for base in complements_seq2:
    mRNA_change2 += mRNA[base]
    


print("sequence2:", sequence1)
print("Complement_seq1:", complements_seq1)
print('mRNA_seq1',mRNA_change1)

print("sequence2:", sequence2)
print("Complement_seq2:", complements_seq2)
print('mRNA_seq2',mRNA_change2)

```

```python

codon_table = {
    'UUU': 'F', 'UUC': 'F',  # Phenylalanine
    'UUA': 'L', 'UUG': 'L',  # Leucine
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',  # Leucine
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I',  # Isoleucine
    'AUG': 'M',  # Methionine (Start)
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',  # Valine

    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',  # Serine
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',  # Proline
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',  # Threonine
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',  # Alanine

    'UAU': 'Y', 'UAC': 'Y',  # Tyrosine
    'UAA': '*', 'UAG': '*', 'UGA': '*',  # Stop codons
    'CAU': 'H', 'CAC': 'H',  # Histidine
    'CAA': 'Q', 'CAG': 'Q',  # Glutamine
    'AAU': 'N', 'AAC': 'N',  # Asparagine
    'AAA': 'K', 'AAG': 'K',  # Lysine
    'GAU': 'D', 'GAC': 'D',  # Aspartic acid
    'GAA': 'E', 'GAG': 'E',  # Glutamic acid

    'UGU': 'C', 'UGC': 'C',  # Cysteine
    'UGG': 'W',  # Tryptophan
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',  # Arginine
    'AGU': 'S', 'AGC': 'S',  # Serine
    'AGA': 'R', 'AGG': 'R',  # Arginine
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'   # Glycine
}


codons_seq1 = [mRNA_change1[i:i+3] for i in range(0, len(mRNA_change1), 3)]
codons_seq2 = [mRNA_change2[i:i+3] for i in range(0, len(mRNA_change2), 3)]

print("mRNA_seq1:", mRNA_change1)
print("Codons_seq1:", codons_seq1)

print("mRNA_seq2:", mRNA_change2)
print("Codons_seq2:", codons_seq2)


protein_seq1 = ""
for codon in codons_seq1:
    protein_seq1 += codon_table.get(codon, "-")  # "-" for unknown codon

print("Protein sequence:", protein_seq1)



protein_seq2 = ""
for codon in codons_seq2:
    protein_seq2 += codon_table.get(codon, "-")  # "-" for unknown codon

print("Protein sequence:", protein_seq2)


```
